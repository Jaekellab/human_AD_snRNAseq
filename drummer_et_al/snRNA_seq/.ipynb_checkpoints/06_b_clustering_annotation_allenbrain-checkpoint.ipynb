## SEA-AD: Map My Cells output

```python
adata = sc.read(DATA_DIR / f"{VERSION}_adata_annotated.h5ad")
adata
```

```python
mapping = pd.read_csv('/lustre/groups/ml01/projects/2023_ad_jaekel/AD1_2025/april22_adata_mapmycells_10xHumanMTGSEA-AD_DeepGenerativeMapping/april22_adata_mapmycells.csv', comment = '#')
mapping.head()
```

- Three Class names:
    - Non-neural
    - Neuronal: Glutamatergic = excitatory neuron
    - Neuronal: GABAergic = inhibitory neuron

```python
class_mapping_dict = {'astro': 'Non-neuronal and Non-neural', 'Endo': 'Non-neuronal and Non-neural', 'Oligo': 'Non-neuronal and Non-neural', 'Micro': 'Non-neuronal and Non-neural', 'OPC': 'Non-neuronal and Non-neural', 'Neuron A': 'Neuronal: Glutamatergic', 'Neuron B': 'Neuronal: GABAergic'}
subclass_mapping_dict = {'astro': 'Astrocyte', 'Endo': 'Endothelial', 'Oligo': 'Oligodendrocyte', 'Micro': 'Microglia-PVM', 'OPC': 'OPC', 'Neuron A': 'Neuronal: Glutamatergic', 'Neuron B': 'Neuronal: GABAergic'}
```

```python
d = {'class_name': np.array(mapping['class_name']), 'subclass_name': np.array(mapping['subclass_name'])}
mapping_df = pd.DataFrame(data = d, index = mapping['cell_id'])
mapping_df.head()
```

```python
def neuron_class_names(row):
    if 'Neuronal' in row['class_name']:
        return row['class_name']
    else:
        return row['subclass_name']
    
mapping_df['subclass_non_neuronal'] = mapping_df.apply(neuron_class_names, axis=1)
mapping_df.head()
```

```python
CELLTYPE_OBS = 'cell_type'
LEIDEN_RES = 'leiden_cluster_1.0'
```

```python
adata_subclass = [subclass_mapping_dict[index] if index in subclass_mapping_dict else index for index in np.array(adata.obs[CELLTYPE_OBS]) ]
adata_class = [class_mapping_dict[index] if index in class_mapping_dict else index for index in np.array(adata.obs[CELLTYPE_OBS]) ]
```

```python
d = {'class_name': adata_class, 'subclass_name': adata_subclass, 'leiden_cluster_0.5': np.array(adata.obs['leiden_res_0.5']), 'leiden_cluster_1.0': np.array(adata.obs['leiden_res_1.0']), 'cortex_area': np.array(adata.obs['brain_area_latin']), 'condition': np.array(adata.obs['condition'])}
adata_df = pd.DataFrame(d, index = adata.obs_names)
adata_df.head()
```

```python
comparison_class = pd.crosstab(adata_df['condition'], adata_df[LEIDEN_RES], rownames = ['condition'], colnames = ['leiden_cluster'], dropna = False, normalize = 'columns')
comparison_class.style.background_gradient(cmap='Blues')
```

```python
comparison_class = pd.crosstab(adata_df['cortex_area'], adata_df[LEIDEN_RES], rownames = ['cortex areas'], colnames = ['leiden_cluster'], dropna = False, normalize = 'columns')
comparison_class.style.background_gradient(cmap='Blues')
```

pd.crossstab():
- normalize:
    - 'index': over each row
    - 'columns': over each column

```python
comparison_class = pd.crosstab(mapping_df['class_name'], adata_df['class_name'], rownames = ['annotationed SEA-AD'], colnames = ['self-annotated'], dropna = False, normalize = False)
comparison_class.style.background_gradient(cmap='Blues')
#dfi.export(styled_df, Path(FIG_DIR, 'table_superclass.png'))
```

```python
comparison_subclass = pd.crosstab(mapping_df['subclass_non_neuronal'], adata_df['subclass_name'], rownames = ['annotationed SEA-AD'], colnames = ['self-annotated'], dropna = False, normalize = False) 
comparison_subclass.style.background_gradient(cmap='Blues')
```

## Distribution of cell types across Leiden clusters according to annotated SEA-AD

ToDO: merge neural excitatory and inhibitory neurons

```python
cell_type_name = 'class_name'

leiden_clusters = adata_df[LEIDEN_RES].unique()
result_class = mapping_df[cell_type_name].groupby(mapping_df[cell_type_name]).apply(lambda x: adata_df[LEIDEN_RES][mapping_df[cell_type_name].isin(x)].value_counts())
class_df = pd.DataFrame(0, index=leiden_clusters, columns = mapping_df[cell_type_name].unique())
for ct in mapping_df[cell_type_name].unique():
    for l, v in result_class[ct].items():
        class_df.loc[str(l), ct] = v
class_df.style.background_gradient(cmap='Blues', axis = "columns")
```

Remove cluster `27` because little cells.

```python
cell_type_name = 'subclass_non_neuronal'

leiden_clusters = adata_df[LEIDEN_RES].unique()
result = mapping_df[cell_type_name].groupby(mapping_df[cell_type_name]).apply(lambda x: adata_df[LEIDEN_RES][mapping_df[cell_type_name].isin(x)].value_counts())
class_df = pd.DataFrame(0, index=leiden_clusters, columns = mapping_df[cell_type_name].unique())
for ct in mapping_df[cell_type_name].unique():
    for l, v in result[ct].items():
        class_df.loc[str(l), ct] = v
class_df.style.background_gradient(cmap='Blues', axis = "columns")
```

```python
# Normalize each row to sum to 1
class_df_normalized = class_df.div(class_df.sum(axis=1), axis=0)

# Sort by index
class_df_normalized = class_df_normalized.sort_index()

# Function to apply background color
def highlight_high_fraction(val):
    return 'background-color: red' if val > 0.7 else ''

# Apply background gradient first
styled_df = class_df_normalized.style.background_gradient(cmap="Blues", axis=1)

# Apply red background where fraction > 0.6
styled_df = styled_df.applymap(highlight_high_fraction)

# Display the styled DataFrame
styled_df
```

```python
for label_key in ["cell_type_sea"]:
    if str(label_key + "_colors") in adata.uns:
        del adata.uns[label_key + "_colors"]
        adata.uns[label_key + "_colors"] = CELL_TYPES_COLOR.values()  # Set custom colours for batches

plt.rcParams['figure.figsize']= (7, 7)
plt.rcParams['figure.figsize']= (7, 7)
for label, colors_dict in [('cell_type_sea',CELL_TYPES_COLOR) , ('condition', CONDITION_COLOR), ('area', AREA_COLOR)]:
    if label + "_colors" in adata.uns:
        del adata.uns[label + "_colors"]
    adata.uns[label + "_colors"] = list(colors_dict.values())
    sc.pl.umap(adata, color=[label], ncols=2, frameon = False, size =7, save=f'_sea_{label}.png')
```
