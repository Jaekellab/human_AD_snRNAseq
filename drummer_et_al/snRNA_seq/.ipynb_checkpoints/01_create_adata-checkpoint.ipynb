Create the `adata` object
----

Loads the count matrix from the `filtered_feature_bc_matrix` in the `cellranger_aggr` results and maps the annotations from the Metadata to the AnnData object.

Output: `AnnData` object saved as `{VERSION}_adata_filtered_feature_matrix.h5ad`

containing
- the following counts
    - raw counts in `adata.X` (will be processed further) and `adata.layers['raw']` (to keep raw counts accessible)
    - spliced, unspliced, and ... counts from `velocyto` in `adata.layers['spliced']`, `adata.layers['unspliced']`, and `adata.layers['...']`
- meta data

## Concatenate count data and metadata files to adata

### Prepare meta data

```python
meta = pd.read_excel(Path(PATH / "Metadata.xlsx"))
meta.loc[meta["Experimental Name sequencing"] == "SJ_2 and SJ_35", "Experimental Name sequencing"] = "SJ_35"
sample_map = pd.read_excel(Path(PATH / "220328_sample_mapping.xlsx"))
sample_map["ExperimentalName"] = sample_map["CaseID_Area_ExperimentalName"].apply(lambda s: s.split("_")[2])

meta["SampleID"] = "-"

for _, sample_info in sample_map.iterrows():
    meta.loc[
        (meta["Experimental Name sequencing"].apply(lambda s: s.split("_")[0]+s.split("_")[1]) == sample_info["ExperimentalName"]),
        "SampleID"
    ] = sample_info["SampleID"]
    

map_brain_areas = {
    'BA17' : 'visual cortex',
    'Area Striata' : 'visual cortex',
    'BA20/21' : 'temporal cortex',
    'Temporal Gyrus' : 'temporal cortex',
    'BA11' : 'prefrontal cortex',
    'BA11/12' : 'prefrontal cortex',
    'Gyrus Rectus' : 'prefrontal cortex',
}    
map_brain_areas_latin = {
    'BA17' : 'Area Striata',
    'BA20/21' : 'Temporal Gyrus',
    'BA11' : 'Gyrus Rectus',
    'BA11/12' : 'Gyrus Rectus',    
    'Area Striata' : 'Area Striata',
    'Temporal Gyrus' : 'Temporal Gyrus',
    'Gyrus Rectus' : 'Gyrus Rectus',
}
map_brain_areas_number = {
    'BA17' : 'BA17',
    'Area Striata' : 'BA17',    
    'BA20/21' : 'BA20/21',
    'Temporal Gyrus' : 'BA20/21',    
    'BA11' : 'BA11',
    'BA11/12' : 'BA11/12',
    'Gyrus Rectus' : 'BA11',    
}


meta["brain area latin"] = meta["Brain area"].apply(lambda s: map_brain_areas_latin[s])
meta["brain area"] = meta["Brain area"].apply(lambda s: map_brain_areas[s])
meta["brain area number"] = meta["Brain area"].apply(lambda s: map_brain_areas_number[s])

meta.head(3)
```

### Load count data

```python
def get_file_path(sample,cr_or_velocyto="cr", matrix_name = "filtered"):
    """Get count data file path for a given sample
    
    sample: str
        Sample str e.g. '22Jan3-DL001'
    cr_or_velocyto: str
        "cr" or "velocyto" for cell ranger h5ad or velocyte loom file
    matrix_name: str
        defines whether to load from 'filtered' or 'raw' matrix
    
    """
    if cr_or_velocyto == "cr":
        spec_path = f"cellranger/{sample}/count_matrices/{matrix_name}_feature_bc_matrix.h5"
        #spec_path = f"scanpy_AnnData/{sample}/{sample}_raw_feature_bc_matrix.h5ad"
    elif cr_or_velocyto == "velocyto":
        spec_path = f"velocyto/{sample}/{sample}.loom"

    return Path(PATH, 'cellranger_results', spec_path)
```

```python
# Define which columns are mapped to obs and under which name
map_meta_cols = {
    "SampleID"                        : "sample",    
    "Disease Condition"               : "condition",
    "brain area"                      : "area",
    "Age"                             : "age",
    "Braak Stage"                     : "Braak stage",
    "Case ID"                         : "donor",
    "Sequencing batch"                : "sequencing batch",
    "Brainbank source"                : "Brainbank source",
    "Experimental Name sequencing"    : "sequencing id",
    "Brain area"                      : None, 
    "Sex"                             : "sex",
    "APOE Genotype"                   : "APOE genotype",
    "brain area latin"                : "brain area latin",
    "brain area number"               : "brain area number",
    "Running number"                  : "running number",
}
```

```python
adatas = []
adatas_raw = []

for i,sample in enumerate(SAMPLES):
    print(get_file_path(sample,cr_or_velocyto="cr"))
    a_cr = sc.read_10x_h5(get_file_path(sample,cr_or_velocyto="cr"))
    a_cr_raw = sc.read_10x_h5(get_file_path(sample,cr_or_velocyto="cr", matrix_name = 'raw'))
    a_vc = sc.read_loom(get_file_path(sample,cr_or_velocyto="velocyto")) # NOTE: Loading a_vc will throw a var_names_not_unique warning, resolved below
    
    # Adjust obs names
    a_cr_obs_names = [o[:-2] for o in a_cr.obs.index]
    a_cr_raw_obs_names = [o[:-2] for o in a_cr_raw.obs.index]
    a_vc_obs_names = [s.split(":")[1][:-1] for s in a_vc.obs.index]
    
    assert [o for o in a_cr.obs.index if o[-2:] != "-1"] == [], "all obs names of a_cr are expected to end with '-1'"
    assert set(a_cr_obs_names) == set(a_vc_obs_names), "obs names of a_cr and a_vc are not identical"
    
    # reindex a_cr.obs and a_vc.obs
    a_cr.obs.index = [o+"-"+str(i) for o in a_cr_obs_names]
    a_cr_raw.obs.index = [o+"-"+str(i) for o in a_cr_raw_obs_names]
    a_vc.obs.index = [o+"-"+str(i) for o in a_vc_obs_names]
    a_vc.obs.index.name = None
    
    # Adjust a_vc.var to make it compatible with a_cr.var
    a_cr.var.rename(columns={'gene_symbols': 'gene_symbol'}, inplace=True)
    a_cr.var_names_make_unique(join='.')
    a_cr_raw.var.rename(columns={'gene_symbols': 'gene_symbol'}, inplace=True)
    a_cr_raw.var_names_make_unique(join='.')
    a_vc.var["gene_symbol"] = a_vc.var.index
    # a_vc.var.set_index('Accession', inplace=True)
    a_vc.var.index.name = None
    a_vc.var_names_make_unique(join='.')

    assert set(a_vc.var_names) == set(a_cr.var_names) == set(a_cr_raw.var_names), "var names of a_cr and a_vc are not identical"
    
    # change obs and var order of a_vc to the same as a_cr
    a_vc = a_vc[a_cr.obs_names,a_cr.var_names]
    
    # combine a_cr and a_vc
    a = ad.AnnData(
        X = a_cr.X,
        obs = a_cr.obs, # a_cr.obs and a_vc.obs are identical
        var = a_vc.var, # a_vc.var has some cool info, a_cr.var not
        layers = a_vc.layers,
    ).copy()

    # combine a_cr and a_vc
    a_raw = ad.AnnData(
        X = a_cr_raw.X,
        obs = a_cr_raw.obs, # a_cr.obs and a_vc.obs are identical
        var = a_vc.var, # a_vc.var has some cool info, a_cr.var not
    ).copy()
    
    # add meta data to obs
    for col_meta, col_obs in map_meta_cols.items():
        if col_obs:
            a.obs[col_obs] = meta.loc[meta["SampleID"]==sample,col_meta].values[0]
            a_raw.obs[col_obs] = meta.loc[meta["SampleID"]==sample,col_meta].values[0]
    
    # add to adatas
    adatas.append(a)
    adatas_raw.append(a_raw)
```

```python
adata = ad.concat(adatas, merge="same")
adata.layers["raw"] = adata.X.copy()
adata.var.index.name = "ensemble_id"
adata.var = adata.var.reset_index().set_index('gene_symbol')
adata.var_names_make_unique()
adata.obs["donor"] = adata.obs["donor"].apply(lambda s: ''.join(str(s).split("/")))
for col in adata.obs.columns:
    if adata.obs[col].dtype == "object":
        adata.obs[col] = adata.obs[col].astype("category")
```

```python
adata_raw = ad.concat(adatas_raw, merge="same")
adata_raw.layers["raw"] = adata_raw.X.copy()
adata_raw.var.index.name = "ensemble_id"
adata_raw.var = adata_raw.var.reset_index().set_index('gene_symbol')
adata_raw.var_names_make_unique()
adata_raw.obs["donor"] = adata_raw.obs["donor"].apply(lambda s: ''.join(str(s).split("/")))
for col in adata_raw.obs.columns:
    if adata_raw.obs[col].dtype == "object":
        adata_raw.obs[col] = adata_raw.obs[col].astype("category")
```

```python
adata_raw.write(Path(DATA_DIR / "adatas" / f"{VERSION}_adata_raw_feature_matrix.h5ad"))
```

## Compare count matrices

Comparision of cellrange and velocyto count matrices in terms of number of counts detected.

Spliced counts from velocyto should closely match cellranger counts.

- Cell ranger counts exonic reads (correspond to mature, spliced mRNA)
- Velocyto's spliced layer includes reads aligning to exons, e.i. spliced transcripts

```python
adata = sc.read(Path(DATA_DIR / "adatas" / f"{VERSION}_adata_filtered_feature_matrix.h5ad"))
adata
```

```python
sns.barplot(
    x=['raw', 'log1p'],
    y=[adata.layers['raw'].sum(), adata.layers['raw'].log1p().sum()]
)
```

```python
sns.barplot(
    x=['cellranger', 'velocyto'],
    y=[adata.layers['raw'].log1p().sum(), adata.layers['velocyto'].log1p().sum()]
)
```

```python
log_counts_per_cell = pd.DataFrame({
    'log_sf': np.concatenate([
        np.array(adata.layers['raw'].log1p().sum(axis=1)).squeeze(), 
        np.array(adata.layers['velocyto'].log1p().sum(axis=1)).squeeze()
    ]),
    'protocol': (['cellranger'] * adata.layers['raw'].shape[0]) + (['velocyto'] * adata.layers['velocyto'].shape[0])
})

sns.histplot(x='log_sf', hue='protocol', data=log_counts_per_cell)
```
