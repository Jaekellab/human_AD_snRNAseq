Number of cells per cluster

```python
# Extract relevant columns
df = adata.obs[['leiden_res_1.0', 'condition', 'scDblFinder_class']]

# Count occurrences per cluster, condition, and scDblFinder_class
value_counts = df.value_counts().reset_index()
value_counts.columns = ['Cluster', 'Condition', 'scDblFinder_class', 'Count']

# Pivot table to get counts of singlets and doublets per Cluster and Condition
pivoted = value_counts.pivot_table(
    index='Cluster',
    columns=['Condition', 'scDblFinder_class'],
    values='Count',
    fill_value=0
)

# Flatten multi-index columns for easier access
pivoted.columns = [f'{cond}_{cls}' for cond, cls in pivoted.columns]
pivoted = pivoted.reset_index()

# Optionally, compute total counts and fractions for singlets only
pivoted['Total_Cells'] = pivoted.filter(like='singlet').sum(axis=1)
pivoted['Fraction_AD'] = pivoted.get('AD_singlet', 0) / pivoted['Total_Cells']
pivoted['Fraction_Ctrl'] = pivoted.get('Ctrl_singlet', 0) / pivoted['Total_Cells']

# Final result
final_table = pivoted[['Cluster', 'Total_Cells', 'Fraction_AD', 'Fraction_Ctrl']]
print(final_table)
```

## Assign Coarse Celltypes to Clusters

```python
LEIDEN_RES = "leiden_res_0.5"
```

```python
sc.pl.dotplot(adata,old_canonical_marker,groupby=LEIDEN_RES)
```

```python
sc.pl.dotplot(adata,CANONICAL_MARKERS,groupby=LEIDEN_RES)
```

Unassigned: 

- 0: GlutaN, Oligo

- 5: GlutaN, Oligo

```python
cell_type_mapping_05 = {
    'Endo': [13],  
    'Micro': [6],
    'GABANeuron': [],
    'GlutaNeuron': [3, 12, 14, 21],
    'Oligo': [1, 7, 10, 11, 17],  
    'OPC': [8],
    'astro': [2, 9, 15, 18, 19] ,
}
```

```python
adata.obs['celltype_05'] = np.nan

for k, v in cell_type_mapping_05.items():
    adata.obs.loc[adata.obs['leiden_res_0.5'].astype(int).isin(v), 'celltype_05'] = str(k)
```
