```python
adata.obsm["nhoods"]
```

```python
nhood_size = adata.obsm["nhoods"].toarray().sum(0)
plt.hist(nhood_size, bins=20)
plt.xlabel("# cells in neighbourhood")
plt.ylabel("# neighbouthoods")
```

```python
np.median(nhood_size)
```

**interpretation**: Median number is good but minimum number of neighbors is not equal to teh proportion in `make_nhoods` (e.i. if prop=0.1 then 10%). Maybe because cell type clusters are quite seperate?

```python
nhood_size = adata.obsm["nhoods"].toarray().sum(0)
plt.hist(nhood_size, bins=20)
plt.xlabel("# cells in neighbourhood")
plt.ylabel("# neighbouthoods")
```

## Cell type proportion t-test

```python
from utils.cell_type_proportions import *
```

```python
# Define covariates to test
covariate_cols = ['condition', 'area']

# Run comprehensive analysis
results = run_comprehensive_analysis_anndata(
    adata, 
    covariate_cols,
    cell_type_col='cell_type_sea',
    sample_col=None,  # Set to your sample column if you have one
    test_type='auto',
    correction_method='fdr',
    alpha=0.05,
    save_results=True,
    save_plots=False
)

# Summary across all covariates
print(f"\n{'='*60}")
print("SUMMARY ACROSS ALL COVARIATES")
print(f"{'='*60}")

for covariate, result_data in results.items():
    results_df = result_data['results_df']
    n_significant = sum(results_df['significant_fdr'])
    total_tests = len(results_df)
    print(f"{covariate}: {n_significant}/{total_tests} significant tests")
```
